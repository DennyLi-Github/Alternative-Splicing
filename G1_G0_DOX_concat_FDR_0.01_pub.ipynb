```python
SE_MATS = pd.read_csv("G1_G0_DOX_concat_38/SE.MATS.JCEC.txt", sep = "\t")
```

```python
SE_MATS
```

```python

SE_MATS_FDR = SE_MATS[(SE_MATS['FDR'] < 0.01) & (SE_MATS['IncLevelDifference'].abs() > 0.25)]
```

```python
SE_MATS_FDR
```

```python
SE_MATS_FDR.to_csv("nease_df_G1_G0_DOX_SE_concact_38_FDR_0.01.csv")
```

```python
SE_MATS_FDR.dtypes
```

```python
# Run NEASE analysis
events = nease.run(
    SE_MATS_FDR,
    organism='Human',  # or 'Mouse'
    input_type='rmats',
    remove_non_in_frame=True,
    only_divisible_by_3=False
)
```

```python
# View summary statistics
stats = events.get_stats()
print(stats)
```

```python
# Get affected domains
# Get list of all protein domais affected by AS.
# A link to DIGGER database for lost/gained domain and their binding.

domains = events.get_domains()
print(domains.head())
```

```python
domains
```

```python
domains.to_csv("domain_G1_G0_DOX_concat_38_FDR_0.01.csv")
```

```python

print(events.enrichment)
```

```python
events.get_edges()
```

```python
events.get_edges().to_csv('edge_G1_G0_DOX_concat_38_FDR_0.01.csv')
```

```python
#Get a list of all affected linear motifs
events.get_elm()
```

```python
events.get_elm().to_csv("elm_G1_G0_DOX_concat_38_FDR_0.01.csv")
```

```python
# get interacting residues affected by AS
events.get_pdb()
```

```python
events.get_pdb().to_csv("pdb_G1_G0_DOX_concat_38_FDR_0.01.csv")
```

```python
pathways = events.enrichment
```

```python
pathways
```

```python
pathways_sorted = pathways.sort_values(by='Nease score', ascending=False)
```

```python
pathways_sorted
```

```python
# Run nease enrichment on affected edges
nease_enr_KEGG=events.enrich(database=['KEGG'])
nease_enr_KEGG.head(10)
```

```python
nease_enr_KEGG.to_csv("SE_nease_enrichment_edge_KEGG_G1_G0_DOX_FDR_0.01.csv")
```

```python
# Run nease enrichment on affected edges
nease_enr_Reactome=events.enrich(database=['Reactome'])
nease_enr_Reactome.head(20)
```

```python
nease_enr_Reactome.to_csv("SE_nease_enrichment_edge_Reactome_G1_G0_DOX_FDR_0.01.csv")
```

```python
#plot nease enrichment for KEGG
nease_enr_KEGG=nease_enr_KEGG.sort_values(by='Nease score',ascending=False)
Term = nease_enr_KEGG['Pathway name'][:10]
Pvalues =nease_enr_KEGG['adj p_value'][:10]
Pvalues=[ -np.log10(x) for x in Pvalues]

plt.barh(Term[::-1],Pvalues[::-1] )
plt.title('NEASE enrichment KEGG')
plt.ylabel('Terms')
plt.xlabel('-log10(Adjusted P-value)')
plt.savefig('SE_nease_enrichment_G1_G0_DOX_edge_KEGG.pdf',format='pdf',bbox_inches='tight')
plt.show()
```

```python
#plot nease enrichment for Reactome
nease_enr_Reactome=nease_enr_Reactome.sort_values(by='Nease score',ascending=False)
Term = nease_enr_Reactome['Pathway name'][:20]
Pvalues =nease_enr_Reactome['adj p_value'][:20]
Pvalues=[ -np.log10(x) for x in Pvalues]

plt.barh(Term[::-1],Pvalues[::-1] )
plt.title('NEASE enrichment Reactome')
plt.ylabel('Terms')
plt.xlabel('-log10(Adjusted P-value)')
plt.savefig('SE_nease_enrichment_G1_G0_DOX_edge_Reactome.pdf',format='pdf',bbox_inches='tight')
plt.show()
```

```python
SE_best_path = events.path_analysis('R-HSA-5653656')
SE_best_path
```

```python
# Extract genes and p-values
Genes = SE_best_path['Spliced genes']
Pvalues = SE_best_path['p_value']

# Convert p-values to -log10 scale
log_pvalues = [-np.log10(p) for p in Pvalues]

Genes = SE_best_path['Spliced genes']
Pvalues = SE_best_path['p_value']
Pvalues=[ -np.log10(x) for x in Pvalues]

# Combine and sort by significance
sorted_data = sorted(zip(Genes, log_pvalues), key=lambda x: x[1], reverse=True)

# Select top N most significant genes
N = 10
top_genes, top_log_pvalues = zip(*sorted_data[:N])


# Plot
plt.figure(figsize=(10, 6))
plt.barh(top_genes, top_log_pvalues)
plt.title('Top 10 Affected Edges Impact on Vesicle-mediated transport')
plt.xlabel('-log10( P-value )')
plt.ylabel('Differentially Spliced Genes')
plt.tight_layout()
plt.savefig('Top_10_SE_nease_G1_G0_DOX_Vesicle-mediated_transport.pdf', format='pdf')
plt.show()
```
